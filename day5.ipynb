```python
class Student:
    def __init__(self,name,age,grade):
        self.name = name
        self.age = age
        self.grade = grade

    def introduce(self):
        print(f"Hi, I am {self.name}. I am {self.age} years old and my grade is {self.grade}.")

student1 = Student("Karthik",20,"A")
student2 = Student("Keerthi",20,"A+")

student1.introduce()
student2.introduce()
```

```python
# Instant variables and methods

class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_item(self,item):
        self.items.append(item)
        print("Item added to cart")
    def show_cart(self):
        if self.items:
            for item in self.items:
                print(f"- {item}")
        else:
            print("Cart is empty")
    def clear_cart(self):
        self.items = []
        print("Cart cleared")


cart = ShoppingCart()
cart.add_item("Helmet")
cart.add_item("Toy car")
cart.add_item("Camera")

cart.show_cart()

cart.clear_cart()

cart.show_cart()
```

```python
#Inheritance and Super()

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hi, I am {self.name} and I am {self.age} old")

class Employee(Person):
    def __init__(self,name,age,company):
        super().__init__(name,age)
        self.company = company
    def show_work(self):
        print(f"I work at {self.company}")

Emp1 = Employee("karthik",20,"EY")
Emp1.introduce()
Emp1.show_work()
```
