```python
def display_student(student):
    for key,value in student.items():
        print(f"{key}: {value}")

student = {
    "name": "Karthik",
    "college": "Raghu Engineering",
    "year": "Final Year",
    "gpa": 8.5
}
display_student(student)
```

```python
def search_by_email_domain(contacts,domain):
    return [name for name,details in contacts.items() if domain in details["email"]]

contacts = {
    "Karthik": {"phone": "9876543210", "email": "karthik@gmail.com"},
    "Rahul": {"phone": "8765432109", "email": "rahul@yahoo.com"},
    "Priya": {"phone": "7654321098", "email": "priya@gmail.com"}
}

domain = "@gmail.com"
print(search_by_email_domain(contacts,domain))
```
