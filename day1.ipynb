```python
def describe_person(name,city):
    print(name,"is From",city)

describe_person("Karthik", "Visakhapatnam")
describe_person("Rahul", "Mumbai")
describe_person("Priya", "Chennai")
```

```python
def calculate_bmi(weight, height):
    return weight / (height ** 2)

my_bmi = calculate_bmi(70,1.75)
print(my_bmi)
```

```python
def describe_course(course, duration = '6 Months'):
    return f"Course: {course} | Duration: {duration}"


print(describe_course("python"))
print(describe_course("python","3 Months"))
```

```python
def multiply_all(*args):
    pr = 1
    for ele in args:
        pr *= ele
    return pr

print(multiply_all(2,3,4))
```

```python
def show_student(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
show_student(name="Karthik", city="Vizag", goal="AI Engineer")
```

```python
def summarize_person(name,*skills,**info):
    print(f"Name: {name}")
    print(f"Skills: {", ".join(skills)}")
    print("Details: ")
    for key,value in info.items():
        print("\t",key,":",value)


summarize_person("Karthik","Python"," Machine Learning", "SQL",college="Raghu Engineering",year="Final Year")
```
